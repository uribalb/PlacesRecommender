```python
import pandas as pd
```

```python
#load dataframe review_restaurants_sentiment from '/content/drive/MyDrive/recommender_dataset/Yelp/yelp_dataset/yelp_review_restaurants_sentiment.csv'

review_restaurants_sentiment = pd.read_csv('/content/drive/MyDrive/recommender_dataset/Yelp/yelp_dataset/yelp_review_restaurants_sentiment.csv')
```

```python
review_restaurants_sentiment.head(5)
```

```python
# Select users who have posted reviews about restaurants
chunksize = 10000  # Adjust based on memory constraints
users_ids = set(review_restaurants_sentiment['user_id'])  # Efficient lookup for filtering
df_users = pd.DataFrame()  # Initialize empty dataframe

for chunk in pd.read_json('/content/drive/MyDrive/recommender_dataset/Yelp/yelp_dataset/yelp_academic_dataset_user.json', lines=True, chunksize=chunksize):
    filtered_chunk = chunk[chunk['user_id'].isin(users_ids)]  # Filter by user_id
    df_users = pd.concat([df_users, filtered_chunk])  # Append filtered chunk

df_users.reset_index(drop=True, inplace=True)  # Reset index for consistency
```

```python
df_users.head(5)
```

```python
#Save newly created users dataframe to the usual location as csv

df_users.to_csv('/content/drive/MyDrive/recommender_dataset/Yelp/yelp_dataset/yelp_users.csv', index=False)
```
